### MATMUL testing

Manually implemented matrix multiplication:

Algorithm:

```
    We need 3 pointers: 
        1. i -> Loop through A rows
        2. j -> Loop through B columns 
        3. k -> Loop through A columns and B rows 
```
Finally build the output(C) tensor: 

$$ C_ij = \sum_k{A_ik * B_kj}$$

````python
B_man = Tensor_CP([
    [1, 2],
    [2, 3],
    [4, 5]
])
B_man.shape
````

````python
B = Tensor([
    [1, 2],
    [2, 3],
    [4, 5]
])
````

````python
data_man.matmul(B_man).data
````

### Checking for the time efficiency of manual and numpy MATMUL

````python
import numpy as np
import timeit

np.random.seed(42)

A_data = np.random.randn(20, 20).astype(np.float32)
B_data = np.random.randn(20, 20).astype(np.float32)

A = Tensor_CP(A_data)
B = Tensor_CP(B_data)

custom_result = A.matmul(B).data
numpy_result = A_data @ B_data

print(np.allclose(custom_result, numpy_result, atol=1e-5))
````

````python
custom_time = timeit.timeit(lambda: A.matmul(B), number=100)
numpy_time = timeit.timeit(lambda: A_data @ B_data, number=100)

print("Custom:", custom_time)
print("NumPy:", numpy_time)
print("NumPy speedup:", custom_time / numpy_time)
````
